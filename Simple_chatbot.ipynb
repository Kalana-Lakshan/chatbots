```python
import random
import re
from datetime import datetime
```

```python
response_dict = {
    r"hello|hi|hey":[
        "Hello! How can I help you today?",
        "Hi there! How can I assist you?",
        "Hey! How may I help you today?"
    ],
    r"how are you?|how do you do?":[
        "I'm doing great.Thank you for asking",
        "I'm fine. How are you?",
        "I'm keeping well. I hope you are also keeping well"
    ]
}

default_responses = [
    "That's interesting! Tell me more",
    "I'm not sure what you are telling. Can you tell me again",
    "Can you rephrase ?"
]
```

```python
def rule_based_chatbot():
  print("chatbot activated")
  while True:
    user_input = input("\nYou: ").strip().lower()
    if user_input in ['quit','exit']:
      print('\nChatbot ended. Thank you!')
      break
    if not user_input:
      print("\nTell me again")
      continue
    response_found = False
    for pattern,responses in response_dict.items():
      if re.search(pattern,user_input,re.IGNORECASE):
        response = random.choice(responses)
        print(f"\n ChatBot: {response}")
        response_found = True
        break
    if not response_found:
        response = random.choice(default_responses)
        print(f"\n ChatBot: {response}")
print("chatbot created successfully!")
rule_based_chatbot()
```
